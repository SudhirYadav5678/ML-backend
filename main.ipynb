```python
housing = strat_train_set.drop("X2 house age",axis=1)
housing_label = strat_train_set["X2 house age"].copy()
```

##for the data missing fields

```python
from sklearn.impute import SimpleImputer

imputer = SimpleImputer(strategy='median')
imputer.fit(housing)
```

```python
imputer.statistics_
```

```python
imputer.statistics_.shape
```

```python
X= imputer.transform(housing)
housing_tr = pd.DataFrame(X, columns=housing.columns)
```

```python
housing_tr.describe()
```

## scikit learn design

there are three types of objects in scikit.
1. Estimater
2. Transformer
3. Predictores

```python
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
my_pipeline = Pipeline([
    ('imputer', SimpleImputer(strategy="median")),
    ('std_scaler', StandardScaler())
])
```

```python
housing_num_tr = my_pipeline.fit_transform(housing_tr)
```

```python
housing_num_tr
```

```python
housing_num_tr.shape
```

## Selecting model for your datasets.

```python
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
#model = LinearRegression()
model = DecisionTreeRegressor()
model.fit(housing_num_tr,housing_label)
```

```python
some_data = housing.iloc[:5]
```

```python
some_label = housing_label.iloc[:5]
```

```python
prepared_data = my_pipeline.transform(some_data)
```

## evaluating model

```python
from sklearn.metrics import mean_squared_error
housing_predctions = model.predict(housing_num_tr)
mse = mean_squared_error(housing_label, housing_predctions)
```

```python
rmse = np.sqrt(mse)
rmse
```

```python
mse
```

cross validation

```python
from sklearn.model_selection import cross_val_score
scores = cross_val_score(model, housing_num_tr, housing_label,scoring="neg_mean_squared_error", cv=10)
rmse_scores = np.sqrt(-scores)
```

```python
rmse_scores
```

```python
def print_scores(scores):
    print("Scores", scores)
    print("mean", scores.mean())
    print("Scores deviation", scores.std())
```

```python
print_scores(rmse_scores)
```
